# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
company_names =  pd.DataFrame( collection_companies.find({"name":"Babelgum"},{'name':1,'_id':0}) )

company_names
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
big_companies =  pd.DataFrame( collection_companies.find( {'number_of_employees':{'$gt':5000}} , 
                                           {'name':1,'_id':0, 'number_of_employees':1} ).limit(20).sort('number_of_employees',-1) )
big_companies #que bonitos son los df. Gracias pandas
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
companies_founded = pd.DataFrame( collection_companies.find({'founded_year':{'$gte':2000,'$lte':2005}},
                                                           {'_id':0,'name':1,'founded_year':1}) )
companies_founded
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
companies_valuation_amount = pd.DataFrame( collection_companies.find( {"ipo.valuation_amount":{'$gt':100000000},"founded_year":{'$lt':2010}},
                                                                     {'_id':0,'name':1,'ipo':1} ) )
companies_valuation_amount
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
small_companies = pd.DataFrame( collection_companies.find( {'number_of_employees':{'$lt':1000},'founded_year':{'$lt':2005}},
                                                         {'_id':0,'name':1,'number_of_employees':1,'founded_year':1}).sort('number_of_employees',-1).limit(10) )
small_companies
```

### 6. All the companies that don't include the `partners` field.

```python
companies_without_partners = pd.DataFrame( collection_companies.find( {'partners':{'$eq':[]}},
                                                                    {'_id':0,'name':1,'partners':1}) )
companies_without_partners
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
companies_null = pd.DataFrame( collection_companies.find( {'category_code':{'$type':'null'}},
                                                         {'_id':0,'name':1,'category_code':1}) )
companies_null
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
medium_companies = pd.DataFrame( collection_companies.find( {'number_of_employees':{'$gte':100, '$lt':1000}},
                                                           {'_id':0,'name':1,'number_of_employees':1}) )
medium_companies
```

### 9. Order all the companies by their IPO price descendently.

```python
ordered_by_ipo = pd.DataFrame( collection_companies.find({},
                                                         {'_id':0,
                                                          'name':1,
                                                          'ipo.valuation_amount':1}).sort('ipo.valuation_amount',-1) ) 
ordered_by_ipo.head(20)
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
biggest_companies = pd.DataFrame( collection_companies.find( {},
                                                       {'_id':0,
                                                        'name':1,
                                                        'number_of_employees':1} ).sort('number_of_employees',-1).limit(10) )
biggest_companies
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
companies_2S = pd.DataFrame( collection_companies.find( {'founded_month':{'$gt':6}},
                                                      {'_id':0,'name':1,'founded_month':1}).limit(1000) )
companies_2S
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
companies_deadpooled = pd.DataFrame( collection_companies.aggregate([
    { '$project': {'_id' : 0, 'name' : 1, 'difference': {'$subtract':["$deadpooled_year","$founded_year"]}}},
    { '$match': {'difference': { '$gt': 3 }}}
]) ) 
companies_deadpooled
```

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
pd.DataFrame( collection_companies.find({},
                                        {'_id':0,
                                        'name':1,
                                       'founded_year':1}).sort('founded_year',1) )
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
pd.DataFrame( collection_companies.find({'founded_day': {'$gte': 0, '$lte': 7}})
             .sort("acquisition.price_amount",-1)
             .limit(10) )
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
pd.DataFrame( collection_companies.find({'category_code':"web",
                                         'number_of_employees':{'$gt':4000}}).sort('number_of_employees',1) )
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.
